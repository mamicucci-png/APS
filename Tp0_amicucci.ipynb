## Generar una señal senoidal

```python
import numpy as np  #importar la libreria numpy

def mi_funcion_sen(vmax, dc,ff,ph,nn,fs): #defino la funcion
   
        ts=1/fs #frecuencia de muestreo
   
        x = np.arange(nn).reshape(-1,1) #-1 rellena automaticamente la cantidad de filas
        
        tt=x*ts #tt=[0,1*ts,2*ts,...,(n-1)*ts]

    #x(t)=Asen(2pi . frecuencia . tiempo +fase) + offset 
    
        xx=vmax * np.sin (2 * np.pi * ff * tt+ ph) + dc
        

        return tt,xx #la funcion devuelve dos resultados tt y xx
    
```

Chequeo que se cumple vector Nx1

```python
vmax=1
dc=0
ff=1 
ph=0 
nn=100
fs=100

tt,xx = mi_funcion_sen(vmax, dc,ff,ph,nn,fs)
print("el tamaño del vector tt es:", tt.shape) #chequeo de vector Nx1
print("el tamaño del vector xx es:",xx.shape) #chequeo de vector Nx1
```

```python
import matplotlib.pyplot as plt #trae la lib de graficos para usar las herramientas
#y la llamamos plt para escribir menos despues
plt.plot(tt, xx) #eje x:tt , eje y : xx
plt.xlabel("Tiempo") #label: nombre a los ejes
plt.ylabel("Voltaje")
plt.title("Senoidal") #titulo al grafico
plt.show() #muestra el grafico en pantalla
```

# Generar una nueva señal
## Señal cuadrada

```python
import numpy as np

def funcion_cuadrada(vmax, dc,ff,ph,nn,fs): 
    ts=1/fs
    
    x = np.arange(nn).reshape(-1,1) #-1 rellena automaticamente la cantidad de filas, comando de python
    tt=x*ts
    
    
    xx = vmax * np.sign(np.sin(2*np.pi*ff*tt +ph))+dc 
    
    return (tt,xx)
```

Gráfico de la nueva señal

```python
vmax=1
dc=0
ff=1 
ph=0 
nn=100
fs=100

tt,xx = funcion_cuadrada(vmax,dc,ff,ph,nn,fs)

import matplotlib.pyplot as plt #trae la lib de graficos para usar las herramientas
#y la llamamos plt para escribir menos despues
plt.plot(tt, xx) #eje x:tt , eje y : xx
plt.xlabel("Tiempo") #label: nombre a los ejes
plt.ylabel("Voltaje")
plt.title("señal cuadrada") #titulo al grafico
plt.show() #muestra el grafico en pantalla
```
